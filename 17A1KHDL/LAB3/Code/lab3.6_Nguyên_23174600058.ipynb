```python
import pandas as pd
```

```python
stocks1 = pd.read_csv(r'D:\17A1KHDL\LAB3\DATA\stocks1.csv')
stocks2 = pd.read_csv(r'D:\17A1KHDL\LAB3\DATA\stocks2.csv')
stocks = pd.concat([stocks1,stocks2],ignore_index=True)
pivot_table = stocks.pivot_table(
    values='close',  
    index='date',    
    columns='symbol',
    aggfunc='mean'
)
pivot_table
```

```python
pivot_table.loc['Tong volume'] = stocks.groupby('symbol')['volume'].sum()
pivot_table
```

```python
pivot_table = pivot_table.sort_values(by='Tong volume',axis=1,ascending=False)
pivot_table
```
